```python
from transformers import pipeline
import pandas as pd
import numpy as np
```

```python
text_classifier = pipeline("sentiment-analysis")
```

```python
from google.colab import userdata
airline_reviews = userdata.get('airline_reviews_dataset')
```

```python
airline_reviews_df = pd.read_csv(airline_reviews)

airline_reviews_df = airline_reviews_df.head(50)
```

```python
airline_reviews_df.shape
```

```python
example = airline_reviews_df['ReviewHeader'][2]
```

```python
text_classifier(example)
```

```python
for index, row in airline_reviews_df.iterrows():
  text = row['ReviewHeader']
  result = text_classifier(text)
  airline_reviews_df.loc[index, 'sentiment'] = result[0]['label']
```
